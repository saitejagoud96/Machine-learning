```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
train=pd.read_csv('titanictrain.csv')
```

```python
test=pd.read_csv('titanictest.csv')
```

```python
train.head(10)
```

```python
test.head()
```

```python
train.isna().any()
```

```python
import math
train['Age']=train['Age'].fillna(int(train.Age.mean()))
```

```python
train.drop(["Name","SibSp","Parch","Ticket",'Cabin',"Embarked"],axis='columns',inplace=True)
```

```python
train.head(10)
```

```python
dummies=pd.get_dummies(train["Sex"])
dummies
```

```python
input1=pd.concat([dummies,train],axis="columns")
input1.head()
```

```python
input1.drop(['Sex'],axis='columns',inplace=True)
```

```python
input1.head()
```

```python
target=input1['Survived']
from sklearn.model_selection import train_test_split
```

```python
input1.drop(['Survived'],axis='columns',inplace=True)
input1.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=45)
```

```python
model.fit(input1,target)
```

```python
model.score(input1,target)
```

```python
test.head()
```

```python
test.drop(['Name',"SibSp","Parch","Ticket","Cabin","Embarked"],axis='columns',inplace=True)
test.head()
```

```python
test.isna().any()
```

```python
test['Age']=test['Age'].fillna(int(test['Age'].mean()))
test['Fare']=test['Fare'].fillna(test['Fare'].mean())
```

```python
test.isna().any()
```

```python
dummies1=pd.get_dummies(test['Sex'])
dummies1
```

```python
test_data=pd.concat([dummies1,test],axis='columns')
test_data.head()
```

```python
test_data.drop(['Sex'],axis='columns',inplace=True)
test_data.head()
```

```python
yp=model.predict(test_data)
yp
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': yp})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
